```python
import pandas as pd

df = pd.read_csv("employees_large.csv")

average_salaries = df.groupby("department")["salary"].mean()
highest_salary_department = average_salaries.idxmax()

print("Average Salary Per Department:")
print(average_salaries)
print("\nDepartment with Highest Average Salary:")
print(highest_salary_department)
```

```python
import pandas as pd

df = pd.read_csv("sales_large.csv")

total_sales = df.groupby(["region", "product"])["sales"].sum()
highest_selling_product = total_sales.idxmax()

print("Total Sales Per Product in Each Region:", total_sales)

print("\nProduct with Highest Total Sales:", highest_selling_product)
```

```python
import pandas as pd
import numpy as np
A = np.array([[2, 4, 6], [1, 3, 5], [7, 9, 11]])
B = np.array([[8, 5, 2], [7, 4, 1], [6, 3, 0]])

sum = A + B
subtraction = A - B
dot_product = np.dot(A, B)
product = A * B

print

print("Matrix Sum:", sum)
print("Matrix Subtraction:", subtraction)
print("Matrix Dot_Product:", dot_product)
print("Matrix Product:", product)
```

```python
import numpy as np

data = np.random.randn(10000)

mean_value = np.mean(data)
median_value = np.median(data)
std_deviation = np.std(data)

min_value = np.min(data)
max_value = np.max(data)

count_above_mean = np.sum(data > mean_value)

normalized_data = (data - min_value) / (max_value - min_value)

print(f"Mean: {mean_value:.4f}, Median: {median_value:.4f}, Std Dev: {std_deviation:.4f}")
print(f"Min: {min_value:.4f}, Max: {max_value:.4f}")
print(f"Count of values > mean: {count_above_mean}")
print("First 5 normalized values:", normalized_data[:5])
```
